    Web scrape the datacamp website to get some useful information

```python
# Import the necessary modules
from bs4 import BeautifulSoup
import requests
import csv

# Web scrape the url below to get the title of each chapter and the description
url = 'https://www.datacamp.com/courses/web-scraping-with-python'

# requests.get() will return a response object and to get the source code from this, we should add .text()
page = requests.get(url).text

soup = BeautifulSoup(page, 'lxml')

# The chapters are contained in the chapters class in the oredered list ol
chapter = soup.find('ol', class_='chapters')

# Store the results in a csv file
csv_file = open('datacamp_courses.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['course_title', 'description'])


# Loop over all chapters to get the title of each and its description
for chap in chapter.find_all('li', class_='chapter'):
    headlines = chap.h4.text
    print(headlines)
    
    description = chap.p.text
    print(description)
    
    
    csv_writer.writerow([headlines, description])

csv_file.close()
```

```python
# Find all datacamp courses, get the title of each and its description
url = 'https://www.datacamp.com/courses/all'

page = requests.get(url).text

soup = BeautifulSoup(page, 'lxml')

courses = soup.find('div', class_='courses__explore-list')

for course in courses.find_all('div', class_='course-block__body'):
    title = course.h4.text
    print(title)
    
    description = course.p.text
    print(description)
    
    print()
```
